```python
x=int(input('Enter a number: '))
y=len(str(x))

while(x>9):
  print(x//10**(y-1), end=', ')
  x%=10**(y-1)
  y-=1

if(x<10):
  print(x, end='')
```

```python
count=0
x=int(input())
y=0

for i in range(1,x+1):
  if(x%i==0):
    y=i+y
    count+=1

if(count==2):
  print(x, 'is a prime number')

elif(count>2 and (y-x)==x):
  print(x, 'is a perfect number')

else:
  print(x, 'is not a prime or perfect number')
```
