```python
x=str(input())
for i in range(len(x)):
  if((x[i]=='0') or (x[i]=='1')):
    z=0
  else:
    z=1
    break
if(z==0):
  print('Binary Number')
else:
  print('Not a Binary Number')
```

```python
x=str(input())
vow='aeiouAEIOU'
con='bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ'
v_count=0
c_count=0
for i in x:
  if i in vow:
    v_count+=1
  if i in con:
    c_count+=1
if((v_count>0 and c_count>0) and (v_count%3==0 and c_count%5==0)):
  print('Aaarr! Me Plunder!!')
else:
  print('Blimey! No Plunder!!')
```
