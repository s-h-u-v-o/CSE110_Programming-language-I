Home Task 1

```python
x=input()
y=''
for i in x :
  if i==y :
    print('',end='')
  else :
    print(i,end='')
    y=i
```

Home Task 2

```python
x=input()
y=0
for i in range(0, len(x)) :
  if x[i]==',' :
    y=i
    break
m=x[0:y]
n=x[y+2:]
if len(m)>=len(n) :
  for i in range(0, len(n)) :
    print(m[i]+n[i], end='')
  print(m[i+1:])
else :
  for i in range(0, len(m)) :
    print(m[i]+n[i], end='')
  print(n[i+1:])
```

```python
x=input()
for i in x :
  if ord(i)==122 :
    print('a', end='')
  elif ord(i)!=122 :
    print(chr(ord(i)+1), end='')
```

Home Task 4

```python
x=input()
count=0
for i in x :
  if 65<=ord(i)<=90 or 97<=ord(i)<=122 :
    if count==0 :
      print(i.upper(), end='')
      count=1
    elif count==1 :
      print(i.lower(), end='')
      count=0
  else :
    print(i, end='')
```
