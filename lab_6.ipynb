```python
x = input()
x = x[1:-1] + ','
y, z = '', []

for i in range(len(x)) :
    if x[i] != ',' :
        y += x[i]
    else :
        z.append(int(y))
        y = ''

print(z)
```

```python
x = input() + ','
y, z = '', []

for i in range(len(x)) :
    if x[i] != ',' :
        y += x[i]
    else :
        z.append(int(y))
        y = ''

if len(z)>3 :
    print(z[2:(len(z)-2)])
else :
    print('Not possible')
```

```python
List_one = [1, 4, 3, 2, 5]
List_two = [8, 7, 6, 9]

for i in range(len(List_one)) :
    if List_one[i] in List_two :
        x = 0
        break
    else :
        x = 1

if x == 0 :
    print(True)
else :
    print(False)
```
