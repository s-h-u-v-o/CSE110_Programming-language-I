```python
x, y, z = input(), [], ''
for i in range(len(x)) :
    if x[i] == ' ' :
        y.append(int(z))
        z = ''
    else :
        z += x[i]
y.append(int(z))
z = [y[i] for i in range(len(y)) if y[i]%2!=0]
print(f'Original list: {y}\nModified list: {z}')
```
