```python
x=int(input('distance: '))
y=int(input('time: '))
z=(x/1000)/(y/3600)

if(z<60):
  print(z,'\n''Too slow. It needs more changes.')

elif(z>=60 and z<=90):
  print(z,'\n''Velocity is okay. The car is ready!')

else:
  print(z,'\n''Too fast. Only a few changes should suffice.')
```
