```python
def is_even(x) :
    if x%2 == 0 :
        return True
    else :
        return False

def is_pos(x) :
    if x >= 0 :
        return True
    else :
        return False

def sequence(x) :
    if is_pos(x) == True :
        for i in range(x+1) :
            if is_even(i) == True :
                print(i, end=' ')
    else :
        for i in range(x, 0) :
            if is_even(i) == False :
                print(i, end=' ')

sequence(-8)
```

```python
def is_valid_triangle(a, b, c) :
    if (a+b > c) and (b+c > a) and (c+a > b) :
        return True
    else :
        return False

def tri_area(a, b, c) :
    if is_valid_triangle(a, b, c) == True :
        s = (a+b+c)/2
        area = (s*(s-a)*(s-b)*(s-c))**(1/2)
        print(round(area, 3))
    else :
        print('Can\'t form triangle')

tri_area(7, 5, 10)
```

CW 3.3

```python
import math

def circle(r) :
    area = math.pi*r**2
    return round(area, 4)

def sphere(r) :
    area = (4/3)*math.pi*r**3
    return round(area, 4)

def fitting(d1, d2, dimension) :
    r1 = d1/2
    r2 = d2/2
    if dimension == 3 :
        vol1 = sphere(r1)
        vol2 = sphere(r2)
        if vol1 > vol2 :
            print(f'Sphere-2 can fit inside Sphere-1 and {round(vol1-vol2, 3)} cubic units would be left.')
        elif vol2 > vol1 :
            print(f'Sphere-1 can fit inside Sphere-2 and {round(vol2-vol1, 3)} cubic units would be left.')
        else :
            print('Impossible to fit.')
    else :
        area1 = circle(r1)
        area2 = circle(r2)
        if area1 > area2 :
            print(f'Circle-2 can fit inside Circle-1 and {round(area1-area2, 3)} square units would be left.')
        elif area2 > area1 :
            print(f'Circle-1 can fit inside Circle-2 and {round(area2-area1, 3)} square units would be left.')
        else :
            print('Impossible to fit.')

fitting(8, 10, 2)

fitting(30, 14, 3)

fitting(5, 5, 3)
```

```python
def is_prime(x) :
    count = 0
    y = 0
    for i in range(1,x+1) :
        if(x%i==0) :
            y=i+y
            count+=1
    if(count==2) :
        return True
    else :
        return False

def is_perfect(x) :
    count = 0
    y = 0
    for i in range(1, x+1) :
        if(x%i==0) :
            y=i+y
            count+=1
    if(count>2 and (y-x)==x) :
        return True
    else :
        return False

def special_sum(x) :
    total_sum = 0
    for i in range(1, x + 1) :
        if is_prime(i) or is_perfect(i) :
            total_sum += i
    return total_sum

x = int(input())
result = special_sum(x)
print(result)
```

```python
def calc_tax(age, salary) :
    if age < 18:
        return 0
    elif salary < 10000 :
        return 0
    elif 10000 <= salary <= 20000 :
        return int(0.07*salary)
    else:
        return int(0.14*salary)

def calc_yearly_tax():
    age = int(input())
    yearly_tax = 0
    for month in range(1, 13):
        income = float(input())
        monthly_tax = float(calc_tax(age, income))
        yearly_tax += monthly_tax
        print(f"Month{month} tax: {monthly_tax}")
    print(f"Total Yearly Tax: {yearly_tax}")

calc_yearly_tax()
```
